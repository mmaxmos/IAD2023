**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
def mn_1():
    o = []
    for i in df['race/ethnicity'].unique():
        sum = 0
        k = 0
        for j in range(0, df.shape[0]):
            if df['race/ethnicity'].iloc[j] == i:
                sum += df['reading score'].iloc[j]
                k += 1
        o.append((i, sum/k))
    return o

def mn_2():
    o =[]
    for i in df['education'].unique():
        mn = 100
        for j in range(0, df.shape[0]):
            if df['education'].iloc[j] == i:
                if mn > df['writing score'].iloc[j]:
                    mn = df['writing score'].iloc[j]
        o.append((i, mn))
    return o

print(mn_1())
print(mn_2())

%timeit mn_1()
%timeit mn_2()
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(av_mark):
    if av_mark >= 90:
        mark = 'A'
    elif av_mark >=80:
        mark = 'B'
    elif av_mark >=70:
        mark = 'C'
    elif av_mark >=60:
        mark = 'D'
    elif av_mark >=50:
        mark = 'E'
    else:
        mark = 'F'
    return mark

#создадим столбец со средним баллом за три экзамена
df['Average mark'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
#создадим столбец с оценками применив функцию GetGrade для перевода баллов в оценку 
df['Grade'] = df['Average mark'].apply(GetGrade)
display(df.head(10))

#выведем кол-во студентов, получивших каждую из оценок
df['Grade'].value_counts().sort_index()
```
